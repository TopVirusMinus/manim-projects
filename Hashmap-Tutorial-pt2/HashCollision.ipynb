```python
from manim import *
config.media_width = '60%'
```

```python
%%manim -v WARNING -ql HashCollison
class HashCollison(Scene):
    def construct(self):
        SQUARE_SIZE = 0.4
        ARRAY_LEN = 8
        RANDOM_NAMES = ["Hamada","Waleed","Marzouk","Kareem","Ashraf","Mustafa","Ahmed","Youssef"]
        
        sq = VGroup(*[Rectangle(width=SQUARE_SIZE*5,height=SQUARE_SIZE, color=DARK_GRAY) for i in range(ARRAY_LEN)]).arrange(DOWN, buff=0.1).to_edge(UR)
        names = VGroup(*[Tex(RANDOM_NAMES[i], color=WHITE, font_size=30).move_to(sq[i].get_center()) for i in range(ARRAY_LEN)])
        idxs =VGroup(*[Tex(i,font_size=25, color=LIGHT_GRAY).move_to(sq[i].get_critical_point(LEFT)).shift(LEFT * 0.2) for i in range(ARRAY_LEN)])
        namesList = VGroup(sq, names, idxs)

        self.add(namesList)
```
